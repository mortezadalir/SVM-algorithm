```python
import numpy as np
import matplotlib.pyplot as plt
from cvxopt import matrix


```

```python
X=np.array([[1,2,2,-1,-2,-2],[0,1,-1,-.5,1,-1]])

Y=np.array([[1],[1],[1],[-1],[-1],[-1]])
c=[]
for item in Y:
    if item==1:
        c.append('red')
    else:
        c.append('blue')
plt.scatter(X[0,:],X[1,:],c=c)
plt.show()
Y
```

```python
X1=2*np.random.randn(100,2)+np.array([2,2])
X2=1.5*np.random.randn(100,2)+np.array([7,6])
data=np.concatenate((X1,X2),axis=0).T
plt.show()
Y=np.concatenate((np.ones(1,0),(-1*np.ones(100))))
c=[]
for item in Y:
    if item==1:
        c.append('red')
    else:
        c.append('blue')
plt.scatter(data[0,:],data[1,:],c=c)
Y
```

```python
X=data
m,n= X.shape
C=1
A1=np.zeros((n,n))
for i in range(n):
    for j in range(n):
        A1[i,j]=np.matmul(X[:,i].T,X[:,j])
A1=A1+(1/C)*np.identity(n)
for i in range(n):
    for j in range(n):
        A1[i,j]=Y[i]*Y[j]*A1[i,j]
Aineq=matrix(-1*np.identity(n))
bineq=matrix(np.zeros((n,1)))
```

```python
f=-1*np.ones((n,1))
Aeq=matrix(Y.T,tc='d')
beq=matrix(np.array([[0]]),tc="d")
beq
```

```python
from cvxopt import solvers
sol = solvers.qp(matrix(A1,tc='d'),matrix(f,tc='d'),Aineq,bineq,Aeq,beq)


```

```python
alpha=np.array(sol['x'])
w=np.zeros((1,2))
for i in range(n):
    w=w+alpha[i]*Y[i]*X[:,i]

w=w.T
```

```python
xx=np.where(alpha.T[0]>.1)

for item in xx[0]:
    if Y[item]<0:
        x_mines=X[:,item]
    if Y[item]>0:
        x_plus=X[:,item]
b=(-0.5)*np.dot(w.T,(x_plus+x_mines))

```

```python
x1=np.linspace(-5,5,200)
x2=-(w[0]/w[1])*x1-b/(w[1])
plt.scatter(X[0,:],X[1,:],c=c)
plt.plot(x1,x2,)
plt.ylim([-5, 5])
plt.show()
```
